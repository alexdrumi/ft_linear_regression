```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [8,4]
```

```python
def quadriatic_function(theta):
    """
    The cost function, J(theta)
    """
    return 0.5*(2*theta-1)**2
```

```python
thetas = np.linspace(-0.2, 1, 10) #evenly spaced numbers over interval [-0.2, 1]
f_vals = quadriatic_function(thetas[:, np.newaxis])

plt.plot(thetas, f_vals)
plt.xlabel('Theta')
plt.ylabel('Objective value')
plt.title('Simple quadratic function')
```
